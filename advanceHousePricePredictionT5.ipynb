## Feature Selection Advanced House Price Prediction

The main aim of this project is to predict the house price
based on various features which we will discuss as we go ahead

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

#for feature selection

from sklearn.linear_model import Lasso
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectFromModel

#to visualise at the columns in the dataframe

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('C:/Users/Lenovo/OneDrive/Documents/train1.csv')
```

```python
dataset.head()
```

```python
# Capture the dependent feature
y_train=dataset[['SalePrice']]
```

```python
## drop dependent feature from dataset
X_train=dataset.drop(['Id','SalePrice'],axis=1)
```

```python
non_numeric_columns= X_train.select_dtypes(include=['object']).columns
print(non_numeric_columns)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in non_numeric_columns:
    X_train[col] = le.fit_transform(X_train[col])
```

```python
X_train.fillna(X_train.mean(), inplace=True)  # Replace with mean (numeric columns)
```

```python
print(X_train.dtypes)
```

```python
'''
Apply Feature Selection
first, I specify the lasso Regression model, and 
select a suitable alpha(equivalent of penalty).
the bigger the aplha the less features that will be selected
then i use the selectFromModel object from sklearn, which
will select the features which coeff are non zero'''


feature_sel_model=SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(X_train,y_train)
```
